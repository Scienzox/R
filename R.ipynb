```r
create_graph <- function(df){
    edges = data.frame(node1 = c("all", "all", "flagged", "flagged", "not flagged", "treated", "not treated"),
                        node2 = c("flagged", "not flagged", "treated", "not treated", "waste_not_flagged", "waste_treated", "waste_not_treated"))
vertices = data.frame(node = c("all", 
                               "flagged", 
                               "not flagged", 
                               "treated", 
                               "not treated", 
                               "waste_not_flagged", 
                               "waste_treated", 
                               "waste_not_treated"),
                               n = c(df$n_tot, 
                                     df$n_flagged, 
                                     df$n_tot - df$n_flagged,
                                     df$n_treated,
                                     df$n_flagged - df$n_treated,
                                     df$waste_not_flagged,
                                     df$waste_treated,
                                     df$waste_not_treated),
                      
                               prop = c(df$n_tot / df$n_tot,
                                        df$n_flagged / df$n_tot,
                                        1.0 -  df$n_flagged / df$n_tot,
                                        df$n_treated / df$n_flagged,
                                        1.0 - df$n_treated / df$n_flagged,
                                        df$waste_not_flagged / (df$n_tot - df$n_flagged),
                                        df$waste_treated / df$n_treated,
                                        df$waste_not_treated / (df$n_flagged - df$n_treated))
                                        
                                    )

    G = graph_from_data_frame(d= edges, 
                               vertices = vertices, 
                               directed = T)
    
    plot(G, 
         layout = layout_as_tree(G),
        vertex.size = V(G)$prop*30,
        edge.label = V(G)$n,
#         node.label = V(G)$prop
        )
    
    return(G)
}

G = create_graph(kpi_df)

V(G)$n
```

# Vectorization

```r
col1 <- runif (12^5, 0, 2)
col2 <- rnorm (12^5, 0, 2)
col3 <- rpois (12^5, 3)
col4 <- rchisq (12^5, 2)
df <- data.frame (col1, col2, col3, col4)
```

No optimization

```r
system.time({
  for (i in 1:nrow(df)) { # for every row
    if ((df[i, 'col1'] + df[i, 'col2'] + df[i, 'col3'] + df[i, 'col4']) > 4) { # check if > 4
      df[i, 5] <- "greater_than_4" # assign 5th column
    } else {
      df[i, 5] <- "lesser_than_4" # assign 5th column
    }
  }
})
```

Optimization

```r
output <- character (nrow(df)) # initialize output vector
system.time({
  for (i in 1:nrow(df)) {
    if ((df[i, 'col1'] + df[i, 'col2'] + df[i, 'col3'] + df[i, 'col4']) > 4) {
      output[i] <- "greater_than_4"
    } else {
      output[i] <- "lesser_than_4"
    }
  }
df$output})
```
